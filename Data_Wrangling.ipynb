```python
import pandas as pd

mydata = pd.read_csv('/content/teleCust1000t.csv', header=None)
#kita kosongkan header, ganti dengan index

mydata.head(10)
```

```python
mydata.columns
```

```python
#nak tukar header/nama columns
name_col = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]

mydata.columns = name_col
mydata.head()
```

**How to deal with missing value**

```python
#to check all the values
mydata.describe(include='all')
```

```python
#display all the columns

pd.set_option('display.max_columns', None)
#pakai formula ni untuk tahu semua columns)

mydata.head(10)
```

```python
mydata.isnull() #check missing value
```

```python
mydata.isnull().sum()
#nak tengok jumlah missing value
#dia tak boleh detect '?' as misssing value, dia boleh trace NaN je
```

```python
#replace the '?' by NaN numpy

import numpy as np

mydata.replace('?', np.nan, inplace=True)

mydata.isnull().sum()
```

```python
#change the missing value of 'normalized-losses)

mean_norm = mydata['normalized-losses'].mean()

mean_norm #ada salah pada datatype ('norm..' as string)
```

```python
mean_norm = mydata['normalized-losses'].astype('float').mean()
#it's always safe to convert to float, rather than int to avoid missing the accuracy

mean_norm
```

```python
mydata['normalized-losses'].replace(np.nan, mean_norm, inplace = True)

mydata.isnull().sum()
```

```python
mydata.describe(include='all') #approach 1: the most top freq
#'four' is categorical data
```

```python
mydata['num-of-doors'].value_counts() #approach2 to check freq
```

```python
mydata['num-of-doors'].value_counts().idxmax()
#approach3 to check the highest freq
#to check the most freq of the data
```

```python
mydata['num-of-doors'].replace(np.nan, mydata['num-of-doors'].value_counts().idxmax, inplace=True)

mydata.isnull().sum()
```

```python
mean_bore = mydata['bore'].astype('float').mean()
#it's always safe to convert to float, rather than int to avoid missing the accuracy

mean_bore
```

```python
mydata['bore'].replace(np.nan, mean_bore, inplace = True)

mydata.isnull().sum()
```

```python
#we only change the 'float' to the mean value while replacing the missing value
mean_stroke = mydata['stroke'].astype('float').mean()
mean_horse = mydata['horsepower'].astype('float').mean()
mean_peak = mydata['peak-rpm'].astype('float').mean()

mydata['stroke'].replace(np.nan, mean_stroke, inplace = True)
mydata['horsepower'].replace(np.nan, mean_horse, inplace = True)
mydata['peak-rpm'].replace(np.nan, mean_peak, inplace = True)

mydata.isnull().sum()
```

```python
#remove the missing value in price
#because we want to make a model to predict the value of the car

mydata.dropna(subset=['price'], axis=0, inplace=True)

mydata.isnull().sum()
```

**Data Formatting**

```python
mydata.dtypes
```

```python
mydata['normalized-losses'] = mydata['normalized-losses'].astype('float')
mydata[['bore', 'stroke', 'horsepower', 'peak-rpm', 'price']] = mydata[['bore', 'stroke', 'horsepower', 'peak-rpm', 'price']].astype('float')

mydata.dtypes
```

```python
mydata.head()
```

```python
#make a new columns, added to the dataset
#the value 235 is the format

mydata['city_L/100KM'] = 235/mydata['city-mpg']
mydata['highway_L/100KM'] = 235/mydata['highway-mpg']

mydata.head()
```

```python
#dropping/deleting the old/unused columns

mydata.drop('city-mpg', axis=1, inplace=True)
mydata.drop('highway-mpg', axis=1, inplace=True)

mydata.head()
```

**Data Normalization**

* Simple Feature Scalling
* Mix-Max
* Z-Score

```python
mydata['width'].head()
```

```python
mydata[['width', 'length', 'height']].head()
```

```python
#Simple feature scalling

mydata['width'] = mydata['width']/mydata['width'].max()

mydata[['width', 'length', 'height']].head()
```

```python
mydata['width'].describe()
```

```python
#min-max

mydata['length'] = (mydata['length'] - mydata['length'].min())/(mydata['length'].max()) - mydata['length'].min()

mydata[['width', 'length', 'height']].head()
```

```python
mydata['length'].describe()
```

```python
#z-score (the most preferred way to normalize data)

mydata['height'] = (mydata['height'] - mydata['height'].mean())/mydata['height'].std()

mydata[['width', 'length', 'height']].head()
```

```python
mydata[['width', 'length', 'height']].describe()
```

**Data Binning**

```python
mydata['price'].describe()
```

```python
price_bins = np.linspace(mydata_new1['price'].min(), mydata_new1['price'].max(), num=4)

print(price_bins)

price_groups = ['Low Price', 'Mid Price', 'High Price']

mydata_new1['Price_bins'] = pd.cut(mydata_new1['price'], bins=price_bins, labels=price_groups, include_lowest=True)

mydata_new1[['price', 'Price_bins']].head(10)
```

```python
mydata_new1['Price_bins'].value_counts()
#the values is not equivalent, so it is imbalance
```

```python
norm_bins = np.linspace(mydata_new1['normalized-losses'].min(), mydata_new1['normalized-losses'].max(), num=4)

print(norm_bins)

norm_groups = ['Low', 'Mid', 'High']

mydata_new1['normalized_losses_bins'] = pd.cut(mydata_new1['normalized-losses'], bins=norm_bins, labels=norm_groups, include_lowest=True)

mydata_new1[['normalized-losses', 'normalized_losses_bins']].head(20)
```
